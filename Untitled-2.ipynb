```python
# Pull 1 year of real Apple price data
aapl = yf.download("AAPL", start="2024-01-01", end="2025-01-01")

print(aapl.tail())
```

```python
# Calculate daily returns
closes = aapl["Close"]
returns = closes.pct_change().dropna()

print(returns.tail())
```

```python
print("Total trading days:", len(returns))
```

```python
# Key financial metrics
avg_return = returns.mean().iloc[0]
volatility = returns.std().iloc[0]
sharpe = (avg_return / volatility) * (252 ** 0.5)

print(f"Average daily return: {avg_return:.4f}")
print(f"Daily volatility: {volatility:.4f}")
print(f"Sharpe ratio: {sharpe:.2f}")
```

```python
stocks = ["AAPL", "MSFT", "GOOGL"]

for stock in stocks:
    data = yf.download(stock, start="2024-01-01", end="2025-01-01", progress=False)
    ret = data["Close"].pct_change().dropna()
    avg = ret.mean().iloc[0]
    vol = ret.std().iloc[0]
    sharpe = (avg / vol) * (252 ** 0.5)
    print(f"{stock} — Sharpe: {sharpe:.2f} | Avg Return: {avg:.4f} | Volatility: {vol:.4f}")
```

```python
import matplotlib.pyplot as plt

# Download data
data = yf.download(["AAPL", "MSFT", "GOOGL"], start="2024-01-01", end="2025-01-01", progress=False)["Close"]

# Normalize to 100 so we can compare them fairly
normalized = data / data.iloc[0] * 100

# Plot
plt.figure(figsize=(12, 6))
plt.plot(normalized["AAPL"], label="AAPL")
plt.plot(normalized["MSFT"], label="MSFT")
plt.plot(normalized["GOOGL"], label="GOOGL")

plt.title("AAPL vs MSFT vs GOOGL — 2024 Performance")
plt.xlabel("Date")
plt.ylabel("Growth of $100")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate returns for all three
returns_all = data.pct_change().dropna()

plt.figure(figsize=(12, 4))

for i, stock in enumerate(["AAPL", "MSFT", "GOOGL"]):
    plt.subplot(1, 3, i+1)
    plt.hist(returns_all[stock], bins=40, edgecolor="black")
    plt.title(f"{stock} Daily Returns")
    plt.xlabel("Return")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Parameters from real AAPL data
start_price = 249.08  # last closing price 2024
avg = 0.0013
vol = 0.0141
days = 252  # simulate 1 year forward
simulations = 1000

# Run simulation
results = []

for i in range(simulations):
    prices = [start_price]
    for d in range(days):
        shock = np.random.normal(avg, vol)
        new_price = prices[-1] * (1 + shock)
        prices.append(new_price)
    results.append(prices)

# Plot
plt.figure(figsize=(12, 6))
for i in range(simulations):
    plt.plot(results[i], alpha=0.05, color="blue")

plt.title("Monte Carlo Simulation — AAPL 1 Year (1000 paths)")
plt.xlabel("Trading Days")
plt.ylabel("Price ($)")
plt.show()
```

```python
# Extract final prices (day 252) from all simulations
final_prices = [results[i][-1] for i in range(simulations)]

# Statistics
final_prices = np.array(final_prices)

print(f"Starting price:       ${start_price:.2f}")
print(f"Average outcome:      ${np.mean(final_prices):.2f}")
print(f"Best case (95th %):   ${np.percentile(final_prices, 95):.2f}")
print(f"Worst case (5th %):   ${np.percentile(final_prices, 5):.2f}")
print(f"Median outcome:       ${np.median(final_prices):.2f}")

# Plot distribution of final prices
plt.figure(figsize=(10, 5))
plt.hist(final_prices, bins=50, edgecolor="black", color="blue", alpha=0.7)
plt.axvline(np.percentile(final_prices, 5), color="red", linestyle="--", label="5th percentile (worst case)")
plt.axvline(np.percentile(final_prices, 95), color="green", linestyle="--", label="95th percentile (best case)")
plt.axvline(np.median(final_prices), color="orange", linestyle="--", label="Median")
plt.title("Distribution of AAPL Prices After 1 Year")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
